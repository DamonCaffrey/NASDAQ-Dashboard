```python
li_check = {}
prev = "ZZZ"

for i in range(df_nasdaq_listings.shape[0]):
    
    curr = df_nasdaq_listings["Symbol"].iloc[i]
    
    if prev == curr[:-1]:
        if prev in li_check.keys():
            li_check[prev].append(curr)
        else:
            li_check[prev] = [prev, curr]
    else:
        prev = curr

li_check
```

```python
s = "INSERT INTO nasdaq VALUES\n"

for i in range(df_nasdaq_listings.shape[0]):
    
    ser = df_nasdaq_listings.iloc[i]
    
    s += f'("{ser[0]}", "{ser[1]}", {ser[2]}, "{ser[3]}", '
    
    if pd.isna(ser[4]):
        s += 'NULL, '
    else:
        s += f'{ser[4]}, '
    
    s += f'"{ser[5]}", "{ser[6]}", '
    
    if pd.isna(ser[7]):
        s += 'NULL)'
    else:
        s += f'{ser[7]})'
    
    if i == df_nasdaq_listings.shape[0] - 1:
        s += ";"
    else:
        s += ",\n"

print(s)
```
